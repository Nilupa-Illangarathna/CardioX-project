```python
# loading the saved model
cardiox_model = pickle.load(open('cardiox_model.sav', 'rb'))
```

```python
@app.post('/ha_prediction')
def predict(input_parameters : model_input):
  input_data = input_parameters.json()
  input_dictionary = json.loads(input_data)

  Novel_datapd=pd.DataFrame.from_dict(input_dictionary,orient='index')
  ND=Novel_datapd.transpose()
  ND=(ND+6.2808752)/13.6829783
  ND=tf.cast(ND,tf.float32)
  threshold=0.02992197871208191
  reconstructions = cardiox_model(ND)
  loss = tf.keras.losses.mae(reconstructions, ND)
  print(str(tf.math.less(loss, threshold).numpy()))
```
